```python
import numpy as np
import matplotlib.pylab as plt
```

## definition of fuzzy numbers
- triangle fuzzy number

```python
# definition of fuzzy numbers
# triangle fuzzy number
class triFuzzy():
    
    def __init__( self, (a1, a2, a3) ): 
        self.borders = np.zeros(3)
        self.borders[0] = float(a1)
        self.borders[1] = float(a2)
        self.borders[2] = float(a3)
        
    # membership function
    def nu( self, x ):
        if len(x) == 1:
            y = 0;
            if ( x < self.borders[0] or x > self.borders[2] ):
                y = 0;
            elif ( x >= self.borders[0] and x <= self.borders[1]):
                y = ( x - self.borders[0] ) / ( self.borders[1] - self.borders[0] )
            elif ( x >= self.borders[1] and x <= self.borders[2] ):
                y = ( self.borders[2] - x ) / ( self.borders[2] - self.borders[1] )
        else:
            y = []
            for i in xrange(len(x)):
                y.append( self.nu( np.array([x[i]]) ) )
            """
            y = x
            c = np.bitwise_or(x < self.borders[0], x > self.borders[2])
            y[c].fill(0)
            c = np.bitwise_and(x >= self.borders[0], x <= self.borders[1])
            y[c] = ( x[c] - self.borders[0] ) / ( self.borders[1] - self.borders[0] )
            c = np.bitwise_and(x >= self.borders[1], x <= self.borders[2])
            y[c] = ( self.borders[2] - x[c] ) / ( self.borders[2] - self.borders[1] )
            """
        return np.array(y)


    # alfa-cut funtion
    # 
    def alfa_cut( self, alfa ):
        a1 = (self.borders[1]-self.borders[0]) * alfa + self.borders[1]
        a2 = (self.borders[2]-self.borders[1]) * alfa + self.borders[2]
        return np.array([a1, a2])
    
    # arithmatic operarions
    def add_tri ( self, B ):
        return triFuzzy(self.borders + B.borders)
    
    def sub_tri ( self, B ):
        return triFuzzy(self.borders - B.borders)
    
    def sym_tri ( self ):
        return triFuzzy(-self.borders)
        
```

```python
A = triFuzzy([-3, 2, 4])
B = triFuzzy([-1, 0, 6])
```

```python
x1 = np.arange(-7, 7, .1)
y1 = A.nu(x1)

x2 = np.arange(-7, 7, .1)
y2 = B.nu(x2)
```

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
C = A.add_tri(B)
x3 = np.arange(-7, 12, .1)
y3 = C.nu(x3)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.show()
```

- Trapezoidal Fuzzy Number

```python
# definition of fuzzy numbers
# Trapezoidal Fuzzy Number
class trapFuzzy():
    
    def __init__( self, (a1, a2, a3, a4) ): 
        self.borders = np.zeros(4)
        self.borders[0] = float(a1)
        self.borders[1] = float(a2)
        self.borders[2] = float(a3)
        self.borders[3] = float(a4)
        
    # membership function
    def nu( self, x ):
        if len(x) == 1:
            y = 0;
            if ( x < self.borders[0] or x > self.borders[3] ):
                y = 0
            elif ( x >= self.borders[0] and x <= self.borders[1]):
                y = ( x - self.borders[0] ) / ( self.borders[1] - self.borders[0] )
            elif ( x >= self.borders[1] and x <= self.borders[2] ):
                y = 1
            elif ( x >= self.borders[2] and x <= self.borders[3] ):
                y = ( self.borders[3] - x ) / ( self.borders[3] - self.borders[2] )
        else:
            y = []
            for i in xrange(len(x)):
                y.append( self.nu( np.array([x[i]]) ) )
            """
            y = x
            c = np.bitwise_or(x < self.borders[0], x > self.borders[2])
            y[c].fill(0)
            c = np.bitwise_and(x >= self.borders[0], x <= self.borders[1])
            y[c] = ( x[c] - self.borders[0] ) / ( self.borders[1] - self.borders[0] )
            c = np.bitwise_and(x >= self.borders[1], x <= self.borders[2])
            y[c] = ( self.borders[2] - x[c] ) / ( self.borders[2] - self.borders[1] )
            """
        return np.array(y)


    # alfa-cut funtion
    # 
    def alfa_cut( self, alfa ):
        a1 = (self.borders[1]-self.borders[0]) * alfa + self.borders[1]
        a2 = (self.borders[2]-self.borders[1]) * alfa + self.borders[2]
        return np.array([a1, a2])
    
    # arithmatic operarions
    def add_trap ( self, B ):
        return trapFuzzy(self.borders + B.borders)
    
    def sub_trap ( self, B ):
        return trapFuzzy(self.borders - B.borders)
    
    def sym_trap ( self ):
        return trapFuzzy(-self.borders)
```

```python
A = trapFuzzy([1, 5, 6, 9])
B = trapFuzzy([2, 3, 5, 8])
```

```python
x1 = np.arange(0, 10, .1)
y1 = A.nu(x1)

x2 = np.arange(0, 10, .1)
y2 = B.nu(x2)
```

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
C = A.add_trap(B)
```

```python
x3 = np.arange(0, 18, .1)
y3 = C.nu(x3)
```

```python
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.show()
```

```python
A = trapFuzzy([-3, 2, 2, 4])
B = trapFuzzy([-1, 0, 0, 6])

x1 = np.arange(-7, 7, .1)
y1 = A.nu(x1)

x2 = np.arange(-7, 7, .1)
y2 = B.nu(x2)

C = A.add_trap(B)
x3 = np.arange(-7, 12, .1)
y3 = C.nu(x3)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.show()
```
